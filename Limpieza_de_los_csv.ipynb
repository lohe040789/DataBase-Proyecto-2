```python
rental['rental_time'] = pd.to_datetime(rental['rental_date']).dt.time
rental.head()
```

```python
rental['date'] = pd.to_datetime(rental['return_date']).dt.date
rental.head()
```

```python
rental['return_time'] = pd.to_datetime(rental['return_date']).dt.time
rental.head()
```

```python
rental = rental.drop(['return_date'], axis=1)
rental.head()
```

```python
rental = rental.rename(columns={'date': 'return_date'})
rental.head()
```
